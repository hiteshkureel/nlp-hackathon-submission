```python
import pandas as pd

train = pd.read_csv("Challenge_1_Disaster_Tweet_Classification/Disaster_with_label.csv")
test = pd.read_csv("Challenge_1_Disaster_Tweet_Classification/Disaster_no_label.csv")

print(train.head())
print(train.columns)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# column ke extra spaces hata do (IMPORTANT)
train.columns = train.columns.str.strip()
test.columns = test.columns.str.strip()

# ab simple naam use kar sakte hain
X = train['Tweet Text']
y = train['label']

vectorizer = TfidfVectorizer(max_features=5000)
X_vec = vectorizer.fit_transform(X)

test_vec = vectorizer.transform(test['Tweet Text'])

model = LogisticRegression()
model.fit(X_vec, y)

preds = model.predict(test_vec)

test['label'] = preds

test.to_csv("submission_challenge1.csv", index=False)

print("DONE ✅")
```

```python
import pandas as pd

train = pd.read_csv(
    "Challenge_2_Fake_News_and_Misinformation_Detection/fakenews_with_labels.csv",
    encoding='latin1',
    on_bad_lines='skip'
)

test = pd.read_csv(
    "Challenge_2_Fake_News_and_Misinformation_Detection/FakeNews_no_labels.csv",
    encoding='latin1',
    on_bad_lines='skip'
)

print(train.head())
print(train.columns)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# clean column names
train.columns = train.columns.str.strip()
test.columns = test.columns.str.strip()

# text NaN fix
train['text'] = train['text'].fillna("")
test['text'] = test['text'].fillna("")

# label clean (IMPORTANT 🔥)
train['label'] = train['label'].astype(str).str.strip()

# sirf valid labels rakho (True/False)
train = train[train['label'].isin(['True', 'False'])]

# features
X = train['text']
y = train['label']

vectorizer = TfidfVectorizer(max_features=5000)
X_vec = vectorizer.fit_transform(X)

test_vec = vectorizer.transform(test['text'])

model = LogisticRegression()
model.fit(X_vec, y)

preds = model.predict(test_vec)

test['label'] = preds

test.to_csv("submission_challenge2.csv", index=False)

print("DONE Challenge 2 ✅")
```
